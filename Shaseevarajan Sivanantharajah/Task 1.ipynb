<h4> Read water quality data

```python
fnamee = "WaterQualityMajor.txt"

watq = pd.read_csv(fnamee,sep='|',encoding='ISO-8859-1', low_memory = False)
```

```python
watq.columns
```

```python
cols = ('StateWellNumber','SampleYear','ParameterCode','ParameterValue')
watqgc = watq.loc[watq["Aquifer"]=="Gulf Coast",cols] 
```

```python
watqgc.head()
```

```python
len(watqgc)
```

```python
watqgc["ParameterCode"].unique()
```

```python
watqgc['EC'] = watqgc.loc[watqgc["ParameterCode"]==94,["ParameterValue"]]
#Create a column of specific conductance
```

```python
watqgc.columns
```

```python
len(watqgc)
```

```python
EC = watqgc.loc[watqgc["ParameterCode"]==94,["ParameterValue"]]
len(EC)
EC
```

```python
watqgc['AS'] = watqgc.loc[watqgc["ParameterCode"]==1000,["ParameterValue"]]
```

```python
AS = watqgc.loc[watqgc["ParameterCode"]==1000,["ParameterValue"]]
len(AS)
```

```python
watqgc
```

```python
watqgcgrp = watqgc.groupby('StateWellNumber')[['StateWellNumber','EC','AS']].mean()
watqgcgrp = watqgcgrp.reset_index(drop=True)
watqgcgrp
```

```python
watqgcmerg = pd.merge(GCdat,watqgcgrp, on="StateWellNumber", how="inner")
watqgcmerg
```

```python
watgECASp = watqgcmerg.dropna(subset=['EC','AS'],how = "all")[['StateWellNumber','LatitudeDD','LongitudeDD','EC','AS']]
```

```python
watgECASp
```

```python
len(watgECASp)
```

```python
watgECASp.to_csv('GCEAS.csv',index=False)
```
